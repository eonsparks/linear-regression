```python
# Required libraries
import matplotlib.pyplot as plt # For visualization
import numpy as np # For calculations
import pandas as pd # For data manipulation
```

```python
# Load dataset
data = pd.read_csv('Salary_Data.csv');
data.head()
```

```python
# Describe data
data.describe()
```

```python
# Relationship between Salary and Experience
plt.scatter(data['YearsExperience'], data['Salary'], color='blue')
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Split variables
x = data.iloc[:, :1].values # Independent variable
y = data.iloc[:, 1:].values # Dependent variable
```

```python
# Compute means of x and y
x_mean = np.mean(x)
print(x_mean)
y_mean = np.mean(y)
print(y_mean)
```

```python
# Compute sum of squares for x and xy
ss_x = np.sum((x - x_mean)**2, axis=0)
ss_xy = np.sum((x - x_mean) * (y - y_mean), axis=0)
```

```python
# Compute β₁(slope) and β₀(intercept)
beta_1 = ss_xy / ss_x
beta_0 = y_mean - (beta_1 * x_mean)
```

```python
# Make Predictions
y_pred = beta_0 + (beta_1 * x)
```

```python
# Model evaluation
residuals = y - y_pred
ss_res = np.sum((y - y_pred) ** 2)
ss_tot = np.sum((y - y_mean) ** 2)

r_squared = 1 - (ss_res / ss_tot)
print(r_squared)
```

```python
# Visual Results
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
